```python
import statistics
age = [24,55,35,26,23,52,25,24,63,31,33,45,59,44,29,21,21,51,42,21]

BMI = [30.2,25.1,35.8,47.9,26.4,35.6,34.3,25.9,32.4,43.3,43.1,30.9,30.1,27.6,41.3,23.2,25.4,36.6,29.3,22.1]

age_mean = statistics.mean(age)

print("age mean", age_mean)

BMI_mean = statistics.mean(BMI)

print("BMI mean", BMI_mean)

age_median = statistics.median(age)

print("age median", age_median)

BMI_median = statistics.median(BMI)

print("BMI median", BMI_median)

age_variance = statistics.variance(age)

print("age variance", age_variance)

BMI_variance = statistics.variance(BMI)

print("BMI variance", BMI_variance)

age_stdev = statistics.stdev(age)

print("age stdev", age_stdev)

BMI_stdev = statistics.stdev(BMI)

print("BMI stdev", BMI_stdev)
```

```python
age = [24,55,35,26,23,52,25,24,63,31,33,45,59,44,29,21,21,51,42,21]

BMI = [30.2,25.1,35.8,47.9,26.4,35.6,34.3,25.9,32.4,43.3,43.1,30.9,30.1,27.6,41.3,23.2,25.4,36.6,29.3,22.1]

age.sort()
BMI.sort()
print(age)
print(BMI)
size = len(age)
print("length of age and BMI: ",size)

age_q1 = age[int(size/4-1)]
print("age Q1:",age_q1)
age_q3 = age[int(3*size/4-1)]
print("age Q3:",age_q3)
print("age IQR:",age_q3-age_q1)
print("age range:",age[-1]-age[1])

BMI_q1 = BMI[int(size/4-1)]
print("BMI Q1:",BMI_q1)
BMI_q3 = BMI[int(3*size/4-1)]
print("BMI Q3:",BMI_q3)
print("BMI IQR:",BMI_q3-BMI_q1)
print("BMI range:",BMI[-1]-BMI[1])

```

```python
from scipy import stats
import numpy as np

age = [24,55,35,26,23,52,25,24,63,31,33,45,59,44,29,21,21,51,42,21]

age = np.array(age)

BMI = [30.2,25.1,35.8,47.9,26.4,35.6,34.3,25.9,32.4,43.3,43.1,30.9,30.1,27.6,41.3,23.2,25.4,36.6,29.3,22.1]

BMI = np.array(BMI)

age_normal = stats.zscore(age)

BMI_normal = stats.zscore(BMI)

# print("age_normal",age_normal)

# print("BMI_normal",BMI_normal)

print("age & BMI \\\\")
for i in range(len(age_normal)):
    print(age_normal[i],' & ',BMI_normal[i],' \\\\')

# calculate Pearson's correlation
corr, _ = stats.pearsonr(age_normal, BMI_normal)
print('Pearsons correlation: %.3f' % corr)

print("covariance",np.cov(age_normal, BMI_normal))
```
